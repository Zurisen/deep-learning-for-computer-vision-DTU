```python
import os
import sys
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
def calc_std(table, col_row):
    std_array = np.array([0])
    for i in range(1,len(table)):
        std_array = np.append(std_array, np.std(table[col_row].values[0:i+1]))
    table["std_"+col_row] = std_array
```

```python
"Call the saved results"
results_folder = "results"
alexnet_adam = pd.read_csv(os.path.join(results_folder, "alexnet_Adam_lr0.001_info.csv"))
alexnet_sgd = pd.read_csv(os.path.join(results_folder, "alexnet_SGD_lr0.001_info.csv"))
vgg11_adam = pd.read_csv(os.path.join(results_folder, "vgg11_Adam_lr0.001_info.csv"))
vgg11_sgd = pd.read_csv(os.path.join(results_folder, "vgg11_SGD_lr0.001_info.csv"))
resnet152_adam = pd.read_csv(os.path.join(results_folder, "resnet152_Adam_lr0.001_info.csv"))
resnet152_sgd = pd.read_csv(os.path.join(results_folder, "resnet152_SGD_lr0.001_info.csv"))
```

```python
"Calculate stds"
calc_std(alexnet_adam, "test_acc")
calc_std(alexnet_sgd, "test_acc")
calc_std(vgg11_adam, "test_acc")
calc_std(vgg11_sgd, "test_acc")
calc_std(resnet152_adam, "test_acc")
calc_std(resnet152_sgd, "test_acc")
```

```python
"Plot accuracies"
matplotlib.rc('xtick', labelsize=15) 
matplotlib.rc('ytick', labelsize=15) 
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize=(25,5))

#plt.figure()
ax1.plot(alexnet_adam.test_acc.values, color="#e40000", label="Adam", linewidth=2)
ax1.plot(alexnet_sgd.test_acc.values, color="#ee6060", label="SGD", linewidth=2)
ax1.fill_between([i for i in range(50)], alexnet_adam.test_acc.values+alexnet_adam.std_test_acc.values, 
                 alexnet_adam.test_acc.values-alexnet_adam.std_test_acc.values,
                 color = "#e40000", edgecolor="#e40000", alpha=0.2)
ax1.fill_between([i for i in range(50)], alexnet_sgd.test_acc.values+alexnet_sgd.std_test_acc.values, 
                 alexnet_sgd.test_acc.values-alexnet_sgd.std_test_acc.values,
                 color = "#ee6060", edgecolor="#ee6060", alpha=0.2)
plt.ylim(min(alexnet_sgd.test_acc.values),1)
ax1.set_title("AlexNet", fontsize=20)
ax1.legend(fontsize=15)

#plt.figure()
ax2.plot(resnet152_adam.test_acc.values, color="#0f00e4", label="Adam", linewidth=2)
ax2.plot(resnet152_sgd.test_acc.values, color="#6960ee", label="SGD", linewidth=2)
ax2.fill_between([i for i in range(50)], resnet152_adam.test_acc.values+resnet152_adam.std_test_acc.values, 
                 resnet152_adam.test_acc.values-resnet152_adam.std_test_acc.values,
                 color = "#0f00e4", edgecolor="#0f00e4", alpha=0.2)
ax2.fill_between([i for i in range(50)], resnet152_sgd.test_acc.values+resnet152_sgd.std_test_acc.values, 
                 resnet152_sgd.test_acc.values-resnet152_sgd.std_test_acc.values,
                 color = "#6960ee", edgecolor="#6960ee", alpha=0.2)
plt.ylim(min(resnet152_sgd.test_acc.values),1)
ax2.set_title("ResNet152", fontsize=20)
ax2.legend(fontsize=15)

#plt.figure()
ax3.plot(vgg11_adam.test_acc.values, color="#00e402", label="Adam", linewidth=2)
ax3.plot(vgg11_sgd.test_acc.values, color="#60ee61", label="SGD", linewidth=2)
ax3.fill_between([i for i in range(50)], vgg11_adam.test_acc.values+vgg11_adam.std_test_acc.values, 
                 vgg11_adam.test_acc.values-vgg11_adam.std_test_acc.values,
                 color = "#00e402", edgecolor="#00e402", alpha=0.2)
ax3.fill_between([i for i in range(50)], vgg11_sgd.test_acc.values+vgg11_sgd.std_test_acc.values, 
                 vgg11_sgd.test_acc.values-vgg11_sgd.std_test_acc.values,
                 color = "#60ee61", edgecolor="#60ee61", alpha=0.2)
plt.ylim(min(vgg11_sgd.test_acc.values),1)
ax3.legend(fontsize=15)
ax3.set_title("VGG-11", fontsize=20)
plt.suptitle("Test Accuracy Curves | l.r. = 0.001", fontsize=20, y=1.1)
plt.savefig("test_acc_curves_lr0.001.svg")
```
